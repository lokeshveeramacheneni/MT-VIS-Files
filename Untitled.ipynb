```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from tqdm import tqdm
```

```python
def read_files(path_name):
    values = []
    files = sorted(glob.glob(os.path.join(path_name+'/*.npy'), recursive=True))
    for filename in files:
        try:
            vals = np.load(filename)
            values.append(vals[0])
        except Exception as e:
            print('Exception: ', e)
            pass
    return np.asarray(values)

def calc_mean_std(path):
    std_vals = []
    mean_vals = []
    for i in tqdm(range(1, 21)):
        path_ = path+str(i)+'/'
        vals = read_files(path_).reshape(-1, 1)
        std_vals.append(np.std(vals))
        mean_vals.append(np.mean(vals))
    return mean_vals, std_vals
```

## Flipout

```python
plt.figure(figsize=(10, 7))
key_, value_, error_ = [], [], []
key_ = range(1, 21)
value_ = fout_sem3d_msp[0]
error_ = fout_sem3d_msp[1]
high_error = []
for i in range(0, len(value_)):
    if value_[i]+error_[i] > 1:
        high_error.append(1-value_[i])
    else:
        high_error.append(error_[i])

for i in range(0, len(value_)):
    if i == 0:
        plt.errorbar(key_[i]-0.2, value_[i], yerr = high_error[i], 
             c='g',capsize=2, elinewidth=3, marker='o', ms=8,
             linestyle='none', label='Flipout-Semantic3D')
    elif i%5 == 0 or i ==19:
        plt.errorbar(key_[i]-1.2, value_[i], yerr = high_error[i], 
                     c='g',capsize=2, elinewidth=3, marker='o', ms=8,
                     linestyle='none')
key_, value_, error_ = [], [], []
key_ = range(1, 21)
value_ = fout_s3dis_msp[0]
error_ = fout_s3dis_msp[1]
high_error = []
for i in range(0, len(value_)):
    if value_[i]+error_[i] > 1:
        high_error.append(1-value_[i])
    else:
        high_error.append(error_[i])
for i in range(0, len(value_)):
    if i == 0 :
        plt.errorbar(key_[i]+0.2, value_[i], yerr = high_error[i], 
             c='r',capsize=2, elinewidth=3, marker='o', ms=8,
             linestyle='none', label='Flipout-S3DIS')
    elif i%5 == 0 or i == 19:
        plt.errorbar(key_[i]-0.8, value_[i], yerr = high_error[i], 
                     c='r',capsize=2, elinewidth=3, marker='o', ms=8,
                     linestyle='none')
plt.grid('equal')
plt.legend(loc='best', fontsize=16)
plt.xticks([1, 5, 10, 15, 20], fontsize=18)
plt.yticks(np.arange(0.55, 1.15, 0.05), fontsize=18)
plt.xlabel('Number of passes', fontsize=18)
plt.ylabel('Mean probability score', fontsize=18)
plt.savefig('Flipout_MSP_cnc.pdf', dpi=600, bbox_inches='tight')
plt.show()
```
